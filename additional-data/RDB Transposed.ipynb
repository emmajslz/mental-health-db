```python
# gdp_transposed = gdp.T
# gdp_transposed = gdp_transposed.reset_index()
# gdp_transposed.to_csv('t_gdp', header=False, index = False)
# Transpose the DataFrame
gdp_transposed = gdp.T

# Set the first row as the column headers and drop the original first row
gdp_transposed.columns = gdp_transposed.iloc[0]  # Set the first row as header
gdp_transposed = gdp_transposed.drop(gdp_transposed.index[0])  # Drop the first row

# Save to CSV without index and without writing additional headers
gdp_transposed.to_csv('t_gdp.csv', index=False)
```

```python

# Function to reshape and save datasets
def reshape_and_save(df, file_name):
    df_melted = pd.melt(df, id_vars=['country'], var_name='year', value_name='value')
    df_melted.to_csv(file_name, index=False)

# Load and reshape 'internet' dataset
internet = pd.read_csv('internet_users.csv')
reshape_and_save(internet, 't_internet.csv')

# Load and reshape 'life_exp' dataset
life_exp = pd.read_csv('lex.csv')
reshape_and_save(life_exp, 't_life_exp.csv')

# Load and reshape 'med_age' dataset
med_age = pd.read_csv('median_age_years.csv')
reshape_and_save(med_age, 't_med_age.csv')

# Load and reshape 'income' dataset
income = pd.read_csv('mincpcap_cppp.csv')
reshape_and_save(income, 't_income.csv')

# Load and reshape 'suicide_count' dataset
suicide_count = pd.read_csv('suicide_total_deaths.csv')
reshape_and_save(suicide_count, 't_suicide_count.csv')

# Load and reshape 'gdp' dataset
gdp = pd.read_csv('gdp_pcap.csv')
reshape_and_save(gdp, 't_gdp.csv')

pop = pd.read_csv('pop.csv')
reshape_and_save(pop, 't_pop.csv')
```
